```python
import requests
my_url = "https://www.goodreads.com/book/most_read"
raw_html = requests.get(my_url).content
```

```python
from bs4 import BeautifulSoup
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

```python
books = []
rows = soup_doc.find_all('tr')

for row in rows:
    title_element = row.find('span', itemprop='name')
    title = title_element.get_text(strip=True) if title_element else "Title not found"
    url_element = row.find('a', class_='bookTitle')
    url = url_element['href'] if url_element else "URL not found"
    img_element = row.find('img', class_='bookCover')
    image_url = img_element['src'] if img_element else "Image not found"
    image_md = "![](" + image_url + ")"
    title_md = "[" + title + "](" + "https://www.goodreads.com" + url + ")"
    books.append({'title': title_md, 'url': 'https://www.goodreads.com' + url, 'image': image_md})

books
```

```python
genres_list = []
for book in books:
    book_url = book['url'] 
    raw_html = requests.get(book_url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    genre_elements = soup_doc.find_all('span', class_='BookPageMetadataSection__genreButton')
    genres = [genre_element.get_text(strip=True) for genre_element in genre_elements]
    book['genres'] = genres 

print(books)
```

```python
import pandas as pd

df = pd.DataFrame(genres_list)

print(df)
```

```python
df.to_csv("books_details.csv", index=False)
```
